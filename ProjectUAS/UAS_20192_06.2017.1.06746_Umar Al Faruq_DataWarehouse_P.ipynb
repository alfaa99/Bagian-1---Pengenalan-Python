```python
import pandas as pd

# read all data
dataset = pd.read_csv('dataset1.csv')
dataset
```

```python
dataset.dtypes
```

```python
#Laporan jurusan yang diambil di universitas

apply_Topic = dataset['Topic'].value_counts().reset_index()
apply_Topic.columns = ['Topic', 'count']
apply_Topic
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.bar(apply_Topic['Topic'], apply_Topic['count'])
plt.xlabel('Topic')
plt.ylabel('Count')
plt.show()
```

```python
#Laporan Rata-Rata Performa Siswa 

raisedhands_mean = dataset['raisedhands'].mean()
VisITedResources_mean = dataset['VisITedResources'].mean()
AnnouncementsView_mean = dataset['AnnouncementsView'].mean()
Discussion_mean = dataset['Discussion'].mean()
contoh_data = [['raisedhands', raisedhands_mean], ['VisITedResources', VisITedResources_mean], ['AnnouncementsView', AnnouncementsView_mean],['Discussion',Discussion_mean]]
contoh_data_frame =  pd.DataFrame(contoh_data, columns=['Performa', 'Rata-Rata'])
contoh_data_frame
```

```python
plt.figure(figsize=(12,8))
plt.bar(contoh_data_frame['Performa'], contoh_data_frame['Rata-Rata'])
plt.title('Rata-Rata Performa Siswa')
plt.xlabel('Performa')
plt.ylabel('Rata-Rata')
plt.show()
```
